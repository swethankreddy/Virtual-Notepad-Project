**Write a function to check whether the number is Palindrome or not.**  
- Hopefully who have did this in c++ in yor CS101

```python
def is_palindrome(number):
    str_number = str(number)
    return str_number == str_number[::-1]
number = int(input("Enter a number: "))
if is_palindrome(number):
    print(f"{number} palindrome")
else:
    print(f"{number}not palindrome")
```


Create a NumPy array of 10 random integers between 1 and 100 and calculate the mean, minimum, and maximum values


```python
import numpy as np
arr = np.random.randint(1,100,size=10)
print("Array :",arr)
print("Mean :",arr.mean())
print("Max :",np.max(arr))
print("Min :",np.min(arr))
```


Create a Pandas DataFrame from a dictionary containing three columns: 'Name', 'Age', and 'City'. Print the DataFrame.
filter the rows where the 'Age' is greater than 30 and print the result.
Add a new column 'Salary' to the DataFrame created in Exercise 1 with values [50000, 60000, 70000]. Print the updated DataFrame
Group the DataFrame by the 'City' column and calculate the average 'Age' for each city. And plot appropriate/required graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Name': ['Jack', 'John', 'Tina', 'David', 'Bob','Lisa'],
    'Age': [24,26,28,39,31,44],
    'City': ['New York', 'New York', 'France', 'France', 'France','Tokyo']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

filtered_df = df[df['Age'] > 30]

print("\nFiltered DataFrame where Age > 30:")
print(filtered_df)

df['Salary'] = [50000,60000,70000,50000,60000,70000]
print("\nDataFrame with Salary:")
print(df)

by_city = df.groupby('City')
print(by_city['Age'].mean())
print(by_city.count())
print(by_city.count()['Name'])
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.show()

plt.bar(df['Name'],df['Salary'])
plt.title('Salaries of People')
plt.xlabel('People')
plt.ylabel('Salaries')
plt.xticks(rotation=45)
plt.show()
plt.pie(by_city.count()['Name'],labels=['France','New York','Tokyo'])
plt.show()
```
